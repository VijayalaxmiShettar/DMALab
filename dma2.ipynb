```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline
```

```python
data1=pd.read_csv("train_users_2.csv")
data2=pd.read_csv("test_users.csv")
```

```python
print(data1.shape)
print(data2.shape)
print(data1.count())
print(data2.count())
```

```python
df_all=pd.concat((data1,data2),axis=0,ignore_index=True)
print(df_all.head())
```

```python
df_all.gender.replace('-unknown-', np.nan, inplace=True)
```

```python
df_all
```

```python
width=0.7
df_all.first_device_type.value_counts(dropna=False).plot(kind='bar',width=width, color='red', rot=0)
plt.xlabel('Devices')
sns.despine()
```

```python
df_all.first_device_type.value_counts(dropna=False)
```

```python
df_all.first_device_type.replace('Mac Desktop', 'Mac', inplace=True)
df_all.first_device_type.replace('Windows Desktop', 'Windows', inplace=True)
```

```python
df_all.signup_app.value_counts(dropna=False)
df_all.signup_app.value_counts(dropna=False).plot(kind='bar',color='blue',rot=0)
plt.xlabel('Sign up App')
sns.despine()

#Most of the accounts are signed up using web
```

```python
df_all['timestamp_first_active']=pd.to_datetime(df_all['timestamp_first_active'],format='%Y%m%d%H%M%S')
df_all['date_account_created']=pd.to_datetime(df_all['date_account_created'],format='%Y-%m-%d')
```
